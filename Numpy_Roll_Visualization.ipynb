```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
from time import sleep
```

```python
# Create a field of linearly varying values
field = np.ones((100, 100))
linear = np.linspace(0, 99, 100)
field *= linear


plt.figure()
plt.imshow(field)
```

```python
# Roll field in +x direction
field_xp = field.copy()
for i in range(101):
  field_xp = np.roll(field_xp, 1, axis=1)
  if i % 5 == 0:
    plt.imshow(field_xp)
    plt.show()
    sleep(0.05)
    clear_output(wait=True)
```

```python
# Roll field in -x direction
field_xn = field.copy()
for i in range(101):
  field_xn = np.roll(field_xn, -1, axis=1)
  if i % 5 == 0:
    plt.imshow(field_xn)
    plt.show()
    sleep(0.05)
    clear_output(wait=True)
```

```python
# Roll field in +y direction
field_yp = field.copy().T
for i in range(101):
  field_yp = np.roll(field_yp, 1, axis=0)
  if i % 5 == 0:
    plt.imshow(field_yp)
    plt.gca().invert_yaxis()
    plt.show()
    sleep(0.05)
    clear_output(wait=True)
```

```python
# Roll field in -y direction
field_yn = field.copy().T
for i in range(101):
  field_yn = np.roll(field_yn, -1, axis=0)
  if i % 5 == 0:
    plt.imshow(field_yn)
    plt.gca().invert_yaxis()
    plt.show()
    sleep(0.05)
    clear_output(wait=True)
```

```python
# Roll field in +x +y direction
field_x = field.copy()
field_y = field.copy().T
field_xp_yp = field_x + field_y
for i in range(101):
  field_xp_yp = np.roll(field_xp_yp, (1,1), axis=(1,0))
  if i % 5 == 0:
    plt.imshow(field_xp_yp)
    plt.gca().invert_yaxis()
    plt.show()
    sleep(0.05)
    clear_output(wait=True)
```

```python
# Roll field in +x -y direction
field_x = field.copy()
field_y = field.copy().T
field_xp_yp = field_x + field_y
for i in range(101):
  field_xp_yp = np.roll(field_xp_yp, (1,-1), axis=(1,0))
  if i % 5 == 0:
    plt.imshow(field_xp_yp)
    plt.gca().invert_yaxis()
    plt.show()
    sleep(0.05)
    clear_output(wait=True)
```
